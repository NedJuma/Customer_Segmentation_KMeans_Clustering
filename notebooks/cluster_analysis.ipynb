# 1) Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

# 2) Load Clustered Data Frame

```python
df=pd.read_csv("/workspaces/Customer_Segmentation_KMeans_Clustering/data/clusters.csv")
df.head()
```

# 3) Cluster Analysis

## a) Analysis by Total Spent

```python
total_spent_summary = (
    df
    .groupby("Cluster")["TotalSpent"]
    .agg(["mean", "median", "min", "max", "count"])
    .round(2)
)

total_spent_summary
```

```python


cluster_means = (
    df
    .groupby("Cluster", as_index=False)["TotalSpent"]
    .mean()
)

fig = px.bar(
    cluster_means,
    x="Cluster",
    y="TotalSpent",
    text_auto=".2f",
    title="Average Total Spending by Customer Cluster"
)

fig.update_layout(
    xaxis_title="Customer Cluster",
    yaxis_title="Average Total Spent ($)"
)

fig.show()
```

### Insights from Total Spent by Cluster

The average total spending varied considerably across the five clusters, indicating that the K-Means algorithm successfully separated customers based on their overall monetary value.

Cluster 2 – High-Value Customers (VIP Segment)
Mean Total Spent: $5,745.16 (Highest)
Median: $2,862.11

This cluster represents the highest-value customers. They contribute the greatest revenue to the business and include the customer with the highest recorded spending ($259,657.30). The large gap between the mean and median indicates the presence of a few exceptionally high spenders within this group. These customers should be prioritized for retention through loyalty programs, exclusive offers, premium customer service, and personalized marketing campaigns.

Cluster 0 – Above-Average Customers
Mean Total Spent: $1,782.40
Median: $1,396.61

Customers in this cluster spend substantially more than the overall average customer but considerably less than those in Cluster 2. The difference between the mean and median suggests moderate variation in spending, with a few higher-spending customers influencing the average. These customers represent an excellent target for upselling and cross-selling strategies aimed at increasing their lifetime value.

Cluster 3 – Moderate-Value Customers
Mean Total Spent: $613.41
Median: $568.24

This is the largest customer segment, comprising 1,441 customers. Spending levels are relatively consistent, as indicated by the small difference between the mean and median. These customers represent the core customer base and provide stable revenue. Promotional campaigns, product recommendations, and bundle offers could encourage them to increase their purchase value.

Cluster 1 – Low-Value Customers
Mean Total Spent: $324.66
Median: $295.24

Customers in this segment have relatively low spending levels. Since the mean and median are close, spending behaviour appears fairly consistent across the cluster. Marketing efforts should focus on increasing engagement through discounts, personalized recommendations, and incentives that encourage repeat purchases.

Cluster 4 – Very Low-Value Customers
Mean Total Spent: $151.50 (Lowest)
Median: $147.28

This cluster contains the lowest-spending customers in the dataset. The close agreement between the mean and median indicates consistently low spending behaviour. These customers may be new shoppers, occasional buyers, or customers with limited purchasing needs. Entry-level promotions, welcome offers, and targeted campaigns designed to encourage a second or third purchase may help increase their long-term value.

Overall Takeaway

The Total Spent analysis reveals a clear hierarchy of customer value across the five clusters. Cluster 2 contains the business's highest-value customers and should be the primary focus of retention strategies, while Cluster 0 consists of promising high-value customers with potential for further growth. Cluster 3 forms the largest and most stable customer base, making it an important target for campaigns aimed at increasing average spending. Finally, Clusters 1 and 4 represent lower-value customers who could benefit from acquisition, activation, and re-engagement campaigns designed to improve their purchasing behaviour.

## e) Analysis by Recency

## Insights from Recency by Cluster

The Recency analysis reveals clear differences in customer activity across the five clusters. Since Recency measures the number of days since a customer's last purchase, lower values indicate more recently active customers, while higher values indicate customers who have not purchased for a longer period.

Cluster 2 – Highly Active Customers
Mean Recency: 22.56 days (Lowest)
Median: 14 days

Cluster 2 consists of the most active customers, having made purchases within the past few weeks on average. Combined with their high spending, high purchase frequency, and high purchasing volume, these customers represent the retailer's most valuable and engaged customer segment. Retaining these customers should be a top priority through loyalty rewards, VIP programs, and personalized product recommendations.

Cluster 3 – Active Regular Customers
Mean Recency: 49.67 days
Median: 36 days

Customers in Cluster 3 have also purchased relatively recently, indicating ongoing engagement with the business. Although they spend less than Cluster 2, they represent the largest customer segment and provide consistent revenue. These customers are ideal candidates for cross-selling, personalized recommendations, and campaigns that encourage slightly more frequent purchases.

Cluster 0 – Moderately Active Customers
Mean Recency: 82.21 days
Median: 63 days

Cluster 0 has not purchased as recently as Clusters 2 and 3 but remains moderately active. Given their high average order value and relatively high total spending, these customers are valuable but may benefit from reminder emails, exclusive promotions, or new product announcements to encourage more frequent purchases.

Cluster 4 – At-Risk Customers
Mean Recency: 90.54 days
Median: 64 days

Customers in Cluster 4 have relatively long periods since their last purchase while also exhibiting the lowest spending and purchasing volume. These customers appear to be drifting away from the business and may require re-engagement campaigns such as discounts, personalized offers, or incentives to return.

Cluster 1 – Dormant Customers
Mean Recency: 271.47 days (Highest)
Median: 269 days

Cluster 1 has the highest recency by a wide margin, indicating that these customers have not made purchases for many months. Combined with their low spending and low purchase frequency, they represent the least engaged customer segment. These customers are at high risk of churn and should be targeted with win-back campaigns, special discounts, or surveys to understand why they have stopped purchasing.

## Full Clusters summary

```python
cluster_summary = (
    df
    .groupby("Cluster")
    .mean(numeric_only=True)
    .reset_index()
)




# Reshape the dataframe
cluster_long = cluster_summary.melt(
    id_vars="Cluster",
    value_vars=[
        "TotalSpent",
        "Orders",
        "ItemsPurchased",
        "AvgOrderValue",
        "Recency"
    ],
    var_name="Feature",
    value_name="Value"
)

fig = px.bar(
    cluster_long,
    x="Cluster",
    y="Value",
    color="Feature",
    barmode="group",
    title="Average Feature Values by Customer Cluster"
)

fig.update_layout(
    xaxis_title="Cluster",
    yaxis_title="Average Value"
)

fig.show()
```
